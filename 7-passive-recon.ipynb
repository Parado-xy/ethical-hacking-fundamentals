### Sub Domain Hunting:
Subdomain hunting is the process of discovering subdomains of a given domain. This can be useful for various purposes, such as identifying potential security vulnerabilities, understanding a website's structure, or even participating in bug bounty programs.

### Techniques for Subdomain Hunting
1. **Google Dorking**: Using specific search operators in Google to find subdomains indexed by Google.
2. **ASN Lookups**: Identifying Autonomous System Numbers (ASNs) associated with an organization to find subdomains.
3. **Subject Alternative Name (SAN) Enumeration**: Using digital certificates to find subdomains protected by a single certificate.
4. **Third-Party Services**: Tools like Amass, DNSDumpster, and VirusTotal can help gather information about subdomains.
5. **Content Security Policy (CSP) Headers**: Analyzing CSP headers to find subdomains.

### Tools for Subdomain Hunting
- **Sublist3r**: A Python-based tool that enumerates subdomains using various search engines and public sources.
- **Amass**: A powerful subdomain discovery tool that performs passive and active reconnaissance.
- **Nmap**: A network scanning tool that can be used to discover subdomains.
- **Knock**: A tool that performs DNS enumeration to find subdomains.
- **Assetfinder**: A tool that searches for subdomains using various sources.
- **crt.sh**: This is a very good website for finding sub domains. It's a certificate search website.

PHP version 7.4.13 has several known vulnerabilities. Some of the key vulnerabilities include:

1. **SQL Injection**: This vulnerability allows attackers to manipulate SQL queries by injecting malicious input.
2. **Cross-Site Scripting (XSS)**: This vulnerability enables attackers to inject malicious scripts into web pages viewed by other users.
3. **Directory Traversal**: This vulnerability allows attackers to access files and directories outside the web root folder.
4. **File Inclusion**: This vulnerability allows attackers to include remote files, potentially leading to remote code execution.
5. **Cross-Site Request Forgery (CSRF)**: This vulnerability allows attackers to perform unauthorized actions on behalf of authenticated users.
6. **Server-Side Request Forgery (SSRF)**: This vulnerability allows attackers to make requests to internal services from the server.
7. **Open Redirect**: This vulnerability allows attackers to redirect users to malicious websites.
8. **Input Validation**: This vulnerability allows attackers to input malicious data that can lead to various attacks.
